```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
species_df = pd.read_csv('species_gene_counts.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(species_df['Gene Count'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Gene Counts')
plt.xlabel('Gene Count')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.savefig('gene_count_histogram.png')  
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.hist(species_df['Gene Count'], bins=30, color='skyblue', edgecolor='black')
plt.yscale('log')  # Set the y-axis to log scale
plt.title('Histogram of Number of Genes per Species in CARD database')
plt.xlabel('Number of Genes per Species')
plt.ylabel('Frequency (Log Scale)')
# plt.grid(True)
plt.tight_layout()
plt.savefig('gene_count_histogram_logscale.png')  
plt.show()
```

```python
top_20_species = species_df.sort_values(by='Gene Count', ascending=False).head(20)
```

```python
import matplotlib.pyplot as plt


top_20_species = top_20_species.drop(top_20_species.index[4])
plt.figure(figsize=(12, 8))
plt.bar(top_20_species['Species'], top_20_species['Gene Count'], color='orange', edgecolor='black')
plt.title('Top 20 Species with Most Genes in CARD database')
plt.xlabel('Species')
plt.ylabel('Number of Genes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('top_20_species_barplot_modified.png')  # Save the bar plot as an image
plt.show()
```
