```python
import pandas as pd

file_path = '/Users/narges/Desktop/Hackthon/card-data/aro_categories.tsv'
aro_categories_df = pd.read_csv(file_path, sep='\t')

print(aro_categories_df.head())

print(aro_categories_df.info())

print(aro_categories_df.describe(include='all'))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
aro_categories_df['ARO Category'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of ARO Categories')
plt.xlabel('ARO Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
